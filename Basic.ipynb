```python
%matplotlib inline
```

```python
# line plot
x = np.linspace(-np.pi,np.pi,200)
C,S = np.cos(x),np.sin(x)
plt.plot(x,C)
plt.plot(x,S)
plt.show()
```

```python
# mutiparameters-setting
# set the lines
plt.plot(x,C,color = 'red', linewidth = 2.0, linestyle = '-')
plt.plot(x,S,color = 'brown', linewidth = 3.0, linestyle = '--')
```

```python
# set the coordinate axis
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(C.min()*1.1,C.max()*1.1)
```

```python
# set the scale of axis

plt.xticks((-np.pi,-np.pi/2,0,np.pi/2,np.pi),(r'$-\pi$',r'$-\pi/2$',r'$+\pi/2$',r'$+\pi$'))
plt.yticks([-1,-0.5,0,0.5,1])
```

```python
# set the form of axis
ax = plt.gca()  #gca: get current axis
ax.spines['right'].set_color('none')  #hide axis
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom') # set the x axis
ax.spines['bottom'].set_position(('data',0)) #set the position of the bottom axis
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

```python
plt.xlim(x.min()*1.1,x.max()*1.1)
plt.ylim(C.min()*1.1,C.max()*1.1)
plt.xticks((-np.pi,-np.pi/2,0,np.pi/2,np.pi),(r'$-\pi$',r'$-\pi/2$',r'$+\pi/2$',r'$+\pi$'))
plt.yticks([-1,-0.5,0,0.5,1])
ax = plt.gca()  #gca: get current axis
ax.spines['right'].set_color('none')  #hide axis
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom') # set the x axis
ax.spines['bottom'].set_position(('data',0)) #set the position of the bottom axis
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

l1, = plt.plot(x,C,color = 'red', linewidth = 2.0, linestyle = '-')
l2, = plt.plot(x,S,color = 'brown', linewidth = 3.0, linestyle = '--')
plt.legend([l1,l2],['cos','sin'],loc = 'upper left')
```

```python
# figure
plt.figure(num = 'sin',figsize=(16,4),dpi = 200)  # num:indentifier  figsize:the size of figure dpi:pixel point
plt.plot(x,S)
plt.figure(num = 'cos',figsize=(20,4),dpi = 200)
plt.plot(x,C)
plt.show()
```

```python
plt.figure(figsize = (20,10))
n = 1024
x = np.random.normal(0,1,n) # generating random numbers normally from 0 to 1
y = np.random.normal(0,1,n) 
T = np.arctan2(y,x)
plt.subplot(1,2,1)
plt.scatter(x,y,s=75,c=T,alpha = .5) #s:the size of scatter  c:color  alpha:
plt.xlim(-1.5,1.5)
plt.xticks(())
plt.yticks(())
plt.ylim(-1.5,1.5)
plt.show()
```

```python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint = True)
Y = np.sin(2*X)
plt.plot(X,Y+1,color = 'blue',alpha = 1.00)
plt.fill_between(X,1.5,Y+1,color = 'blue', alpha = .25)  # 1.5: y=1.5

plt.show()
```

```python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint = True)
Y = np.sin(2*X)
plt.plot(X,Y+1,color = 'blue',alpha = 1.00)
plt.fill_between(X,Y+1,(Y-1)>-1,color = 'blue', alpha = .25)
plt.fill_between(X,Y+1,(Y-1)<-1,facecolor = 'blue', alpha = .25)
plt.show()
```

```python
n = 12
x = np.arange(12)
y1 = (1-x/float(n))*np.random.uniform(0.5,1.0,n)
y2 = (1-x/float(n))*np.random.uniform(0.5,1.0,n)
plt.bar(x,+y1,facecolor = 'red',edgecolor = 'white',alpha = 0.7)
plt.bar(x,-y2,facecolor = 'blue',edgecolor = 'black')
plt.xticks(())
plt.yticks(())
for x1,y in zip(x,y1):
    plt.text(x1-0.5,y+0.05,'%.2f'%y,color='blue')
for x2,y in zip(x,y2):
    plt.text(x2-0.5,-y-0.07,'-%.2f'%y,color='red')
plt.show()
```
